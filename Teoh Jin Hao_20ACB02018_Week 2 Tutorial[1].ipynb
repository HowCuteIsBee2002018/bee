```python
#functions that can compare 2 numbers
def compare(x, y):
    """this function is to compare 2 numbers"""
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is equal to {y}")
```

```python
compare(9, 0)
```

## create custom images

```python
import matplotlib.pyplot as plt
```

```python
#create a black grayscale image
black= np.zeros((200,100), dtype=np.uint8) # "unit8"

plt.imshow(black, cmap=plt.cm.gray) #"gray"
plt.title("black image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#create a white grayscale image
white= np.zeros((200,100)) + 255
white= np.uint8(white)
# white= white.astype("uint8")
plt.imshow(white, cmap=plt.cm.gray, vmin=0, vmax=256) #"gray"
plt.title("white image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#random color image
random_color_img= np.random.randint(0, high= 256, size= (150, 250, 3), dtype=np.uint8)

plt.imshow(random_color_img)
plt.title("random color image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

def oddeven(x):
    if isinstance(x, int):
        if x%2==0:
            if x==0:
                ans=" is neither odd or even."
            else:
                ans=" is even number."
        elif x%2==1:
            ans=" is odd number."
    else:
        ans="this is not an integer."
    return print(f"{x}" + ans)
```

```python
oddeven(0)
```

## exercise 4

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

# Create a figure and axes
fig, ax = plt.subplots()

# Create a black square patch
square = Rectangle((0, 0), 1, 1, facecolor='black')

# Create a white circle patch
circle = Circle((0.5, 0.5), 0.1, edgecolor='black', facecolor='white')

# Add the square and circle patches to the axes
ax.add_patch(square)
ax.add_patch(circle)

# Set the aspect ratio to 'equal' to ensure the square and circle are not distorted
ax.set_aspect('equal')

# Set the x and y limits of the plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Display the plot
plt.xticks([]), plt.yticks([])
plt.show()
```
